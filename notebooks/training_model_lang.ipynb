```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv("/Users/nidhishgupta/Desktop/Movie_recommendation_sysytem/data/processed/processed_language/reduce_processed_data_lang.csv")
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else []
)
# movies['overview']= movies['overview'].apply(lambda x: " ".join(x))
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: " ".join(x))
```

```python
movies['overview']
```

```python
movies['director'] = movies['director'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else []
)
movies['director'] = movies['director'].apply(lambda x: " ".join(x))
movies['director']
```

```python
movies['cast'] = movies['cast'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else []
)
movies['cast'] = movies['cast'].apply(lambda x: " ".join(x))
movies['cast']
```

```python
movies['genres'] = movies['genres'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else []
)
movies['genres'] = movies['genres'].apply(lambda x: " ".join(x))
movies['genres']
```

```python
movies['production_companies'] = movies['production_companies'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else []
)
movies['production_companies'] = movies['production_companies'].apply(lambda x: " ".join(x))
movies['production_companies']
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
tfidf_overview = TfidfVectorizer(
    max_features=15000,      # controls memory
    stop_words='english',    # removes common words
    min_df=3                 # ignore very rare words
)

X_overview = tfidf_overview.fit_transform(movies['overview'])
```

```python
cv_genres = CountVectorizer()

X_genres = cv_genres.fit_transform(movies['genres'])
```

```python
cv_cast = CountVectorizer(
    max_features=5000    # important to control size
)

X_cast = cv_cast.fit_transform(movies['cast'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
lang_encoder = OneHotEncoder(
    sparse_output=True,
    handle_unknown="ignore"
)
```

```python
X_lang = lang_encoder.fit_transform(
    movies[['original_language']]
)
```

```python
import joblib

joblib.dump(
    lang_encoder,
    "/Users/nidhishgupta/Desktop/Movie_recommendation_sysytem/models/models_including_lang/language_encoder.pkl"
)
```

```python
cv_director = CountVectorizer(
    max_features=5000    # important to control size
)

X_director = cv_director.fit_transform(movies['director'])
```

```python
cv_prod = CountVectorizer(
    max_features=2500    # important to control size
)

X_prod = cv_prod.fit_transform(movies['production_companies'])
```

```python
from scipy.sparse import hstack
```

```python
W_OVERVIEW = 0.4
W_GENRES   = 1.8
W_CAST     = 1.6
W_DIRECTOR = 0.8
W_PROD = 0.6
W_LANG = 2.0
```

```python
X_combined = hstack([
    W_OVERVIEW * X_overview,
    W_GENRES   * X_genres,
    W_CAST     * X_cast,
    W_DIRECTOR * X_director,
    W_PROD * X_prod,
    W_LANG * X_lang
])
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
N_COMPONENTS = 400
```

```python
svd = TruncatedSVD(
    n_components=N_COMPONENTS,
    random_state=42
)

X_reduced = svd.fit_transform(X_combined)
```

```python
X_reduced = X_reduced.astype('float32')

# Normalize for cosine similarity
X_reduced /= np.linalg.norm(X_reduced, axis=1, keepdims=True)
```

```python
import faiss

d = X_reduced.shape[1]  # number of dimensions

index = faiss.IndexFlatIP(d)  # Inner Product
index.add(X_reduced)

print("Total vectors indexed:", index.ntotal)
```
